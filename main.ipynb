```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/Users/gracewang/Documents/aptos-dr-classification/data/train.csv")
print(df.columns)
```

```python
# First split: train vs temp
x_train, x_temp, y_train, y_temp = train_test_split(
    df["id_code"],
    df["diagnosis"],
    stratify=df["diagnosis"],
    test_size=0.30,
    random_state=1
)

# Second split: validation vs test
x_val, x_test, y_val, y_test = train_test_split(
    x_temp,
    y_temp,
    stratify=y_temp,
    test_size=0.50,
    random_state=1
)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

y_train.value_counts().sort_index().plot(kind="bar", ax=axes[0], title="Train")
y_val.value_counts().sort_index().plot(kind="bar", ax=axes[1], title="Validation")
y_test.value_counts().sort_index().plot(kind="bar", ax=axes[2], title="Test")

for ax in axes:
    ax.set_xlabel("DR Severity")
    ax.set_ylabel("Count")

plt.tight_layout()
plt.show()
```
